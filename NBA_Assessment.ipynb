**Part I: Data Ingestion, Data Manipulation, and Data Export**
- The attached CSV file data.csv has the Home_Score and Away_Score for every event that occurred in an NBA game over the past 5 regular seasons. 

- Please create a new column in the CSV file called Score_Margin using the below formula:
Score_Margin = ABS(Home_Score – Away_Score)
where “ABS” = Absolute Value

- Export the data with the new column Score_Margin to a CSV file called new_data.csv. Your file new_data.csv should contain 6 columns: Season, Game, Event_Num, Home_Score, Away_Score, and Score_Margin.

```python
import pandas as pd
```

```python
# Read the csv file and print the first five rows

df = pd.read_csv("data.csv")
df
```

```python
# Create a new column "Score_Margin" with the absolute value of the difference
# between "Home_Score" and "Away_Score"

df['Score_Margin']=(df['Home_Score']-df['Away_Score']).abs()

df
```

```python
# Creation of a CSV file 

df.to_csv("new_data.csv",index=False)
```

**Part II: Data Manipulation and Analysis**

- Using the new column Score_Margin, please find the final score margin for each game in the 2020-21 regular season (hint: the final score margin should be calculated using the last row of each game, so there should only be 1 row per game to make the below calculations).

```python
# Select the "Game" and "Score_Margin" columns during the season 2020-21

df = df.loc[df['Season'] == '2020-21']
df = df[['Season','Game','Event_Num','Score_Margin']]
df.sort_values(['Game','Event_Num'])
```

There were 1080 games during the season 2020-21.
This number does not seem realistic in my opinion.


```python
# Group the dataframe by game and select only the last row of each game
# Sort the game by ascending value and rectify the index

df = df.groupby('Game')
df = pd.concat([df.tail(1)])
df = df.sort_values(['Game','Event_Num'])
df = df.reset_index(drop=True)
df
```
